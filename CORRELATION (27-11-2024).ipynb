```python
import pandas as pd
data = pd.read_csv('/content/titanic/train_and_test2.csv')

data.head()
```

```python
# prompt: give command for matplot-different graphs

import matplotlib.pyplot as plt
import seaborn as sns

# Histogram
plt.figure(figsize=(8, 6))
plt.hist(data['Age'], bins=20, color='skyblue', edgecolor='black')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()


# Scatter plot
plt.figure(figsize=(8, 6))
plt.scatter(data['Age'], data['Fare'], color='green', alpha=0.7)
plt.title('Age vs. Fare')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.show()


# Box plot
plt.figure(figsize=(8, 6))
sns.boxplot(x='Pclass', y='Age', data=data, palette='pastel')
plt.title('Age Distribution by Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Age')
plt.show()


# Bar chart
survived_counts = data['2urvived'].value_counts()
plt.figure(figsize=(6, 4))
survived_counts.plot(kind='bar', color=['salmon', 'lightgreen'])
plt.title('Survival Counts')
plt.xlabel('Survived')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.show()


# Pie chart
plt.figure(figsize=(6, 6))
data['Sex'].value_counts().plot.pie(autopct='%1.1f%%', startangle=90, colors=['lightblue', 'lightcoral'])
plt.title('Sex Distribution')
plt.ylabel('')  # Hide y-label
plt.show()


# Violin plot
plt.figure(figsize=(8, 6))
sns.violinplot(x="Sex", y="Age", hue="2urvived", data=data, split=True, inner="quart")
plt.title('Age Distribution by Sex and Survival')
plt.show()
```

```python
# prompt: give command for seaborn-different graphs

# Heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(data.corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix')
plt.show()

# Pair plot
sns.pairplot(data[['Age', 'Fare', 'Pclass', '2urvived']], hue='2urvived', diag_kind='kde')
plt.show()

# Joint plot
sns.jointplot(x='Age', y='Fare', data=data, kind='kde', color='purple')
plt.show()

# Count plot
plt.figure(figsize=(8, 6))
sns.countplot(x='Embarked', hue='2urvived', data=data, palette='Set2')
plt.title('Embarked Location vs Survival')
plt.show()

# Swarm plot
plt.figure(figsize=(10,6))
sns.swarmplot(x='Pclass', y='Age', hue='2urvived', data=data, dodge=True, palette='Set1')
plt.show()

# KDE Plot
plt.figure(figsize=(8, 6))
sns.kdeplot(data=data['Age'], shade=True, color='orange')
plt.title('KDE Plot of Age')
plt.show()

```

```python
# prompt: give code for binomial distribution graphs

import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import binom

# Parameters for the binomial distribution
n = 10  # Number of trials
p = 0.5  # Probability of success

# Generate x values (number of successes)
x = np.arange(0, n + 1)

# Calculate the probability mass function (PMF)
pmf = binom.pmf(x, n, p)

# Create the plot
plt.figure(figsize=(8, 6))
plt.bar(x, pmf, label='Binomial PMF')
plt.xlabel('Number of Successes')
plt.ylabel('Probability')
plt.title(f'Binomial Distribution (n={n}, p={p})')
plt.xticks(x)  # Ensure integer ticks on x-axis
plt.legend()
plt.grid(True)
plt.show()


# Cumulative Distribution Function (CDF)
cdf = binom.cdf(x, n, p)
plt.figure(figsize=(8, 6))
plt.plot(x, cdf, marker='o', linestyle='-', label='Binomial CDF')
plt.xlabel('Number of Successes')
plt.ylabel('Cumulative Probability')
plt.title(f'Binomial Cumulative Distribution (n={n}, p={p})')
plt.xticks(x)
plt.legend()
plt.grid(True)
plt.show()
```

```python


import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming 'data' DataFrame is already loaded as in the previous example

# Calculate the correlation matrix using Spearman rank correlation
correlation_matrix_spearman = data.corr(method='spearman')

# Create the heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix_spearman, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Spearman Rank Correlation Matrix')
plt.show()

# Calculate the correlation matrix using Kendall Tau correlation
correlation_matrix_kendall = data.corr(method='kendall')

# Create the heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix_kendall, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Kendall Tau Correlation Matrix')
plt.show()
```
