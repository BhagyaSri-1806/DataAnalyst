# Distribution plots


Distribution plots are charts that show how data is spread out. They help you see patterns in your data, like where most values are, how spread out they are, and whether there are unusual values (outliers).

**1. Histogram**

What it is: A bar chart that shows how many data points fall into specific ranges (called bins).
Why it’s useful: It helps you see the shape of your data (e.g., normal, skewed, or uniform).

**2. Box Plot**

What it is: A simple chart that shows the minimum, maximum, median, and spread of data, and highlights any outliers.
Why it’s useful: It’s great for spotting outliers and understanding the overall range of the data

**3. KDE Plot (Smooth Curve)**

What it is: A smooth line that shows the probability density of your data (like a smoothed histogram).
Why it’s useful: It helps you see the overall shape of the data more clearly than a histogram.

**4. Combined Histogram and KDE Plot**

What it is: A histogram with a KDE curve overlaid on top.
Why it’s useful: Combines the details of a histogram with the smooth visualization of a KDE.

# **4. Timestamp Transformation**

# 1.	Convert the Order Date column to a datetime format.

```python
# prompt:  Code Convert the Order Date column to a datetime format.

# Convert the 'order_date' column to datetime format
data['Date'] = pd.to_datetime(data['Date'], errors='coerce')

# Now you can work with the 'order_date' column as datetime objects.
# For example, you can extract the year, month, day, etc.
data['order_year'] = data['Date'].dt.year
data['order_month'] = data['Date'].dt.month
data['order_day'] = data['Date'].dt.day

# Display the updated DataFrame with the new datetime features
print(data.head())
```

# 2.	Create new columns:
o	Day of the week.

o	Day of the month.

o	Month.

o	Year.


o	Season (Winter, Spring, Summer, Fall).


```python
# prompt: 2.	Create new columns:
# o	Day of the week.
# o	Day of the month.
# o	Month.
# o	Year.
# o	Season (Winter, Spring, Summer, Fall).

# Assuming 'order_date' column exists and is in datetime format (as per previous code)

# Extract day of the week (Monday=0, Sunday=6)
data['day_of_week'] = data['Date'].dt.dayofweek

# Extract day of the month
data['day_of_month'] = data['Date'].dt.day

# Extract month
data['month'] = data['Date'].dt.month

# Extract year
data['year'] = data['Date'].dt.year

# Define a function to determine the season
def get_season(month):
    if month in [12, 1, 2]:
        return 'Winter'
    elif month in [3, 4, 5]:
        return 'Spring'
    elif month in [6, 7, 8]:
        return 'Summer'
    else:
        return 'Fall'

# Apply the function to create the 'season' column
data['season'] = data['month'].apply(get_season)

# Display the updated DataFrame
print(data.head())
```
