# **1. Business Problem**

# -->The bakery wants to increase sales and optimize inventory. Currently, they face challenges in:

# -->Understanding peak selling times (e.g., morning vs. afternoon).
#--> Identifying best-selling and least-selling items to improve stock management.
# -->Predicting demand to reduce waste and avoid stockouts.



# **2. Business Objective**
# The goal is to analyze customer purchasing patterns to improve business operations:

# -->Increase sales by offering promotions during low-sales periods.

# -->Introduce new products based on customer preferences.

# **3. Business Constraints**
# Some sales might not be recorded correctly.
# If the time of purchase is wrong, we may misinterpret peak hours.
# "Choco Cake" and "Chocolate Cake" might be counted as different

```python


import pandas as pd
import matplotlib.pyplot as plt

# Assuming 'data' DataFrame is already loaded as in the previous example

# Calculate item frequencies
item_counts = data['Items'].value_counts()

# Get the top 10 items
top_10_items = item_counts.head(10)

# Create a bar chart
plt.figure(figsize=(10, 6))
top_10_items.plot(kind='bar')
plt.title('Top 10 Best-Selling Items')
plt.xlabel('Items')
plt.ylabel('Number of Sales')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

# Create a pie chart
plt.figure(figsize=(8, 8))
plt.pie(top_10_items, labels=top_10_items.index, autopct='%1.1f%%', startangle=90)
plt.title('Top 10 Best-Selling Items (Percentage)')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```
