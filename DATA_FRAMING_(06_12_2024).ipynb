```python
#loc

# .loc is primarily label based, but may also be used with a boolean array.
# .loc will raise KeyError when the items are not found.

# Allowed inputs are:
# A single label, e.g. 5 or 'a', (note that 5 is interpreted as a label of the index, and never as an integer position along the index).
# A list or array of labels, e.g. ['a', 'b', 'c'].
# A slice object with labels, e.g. 'a':'f'.
# A boolean array of the same length as the axis being sliced, e.g. [True, False, True].
# A callable function with one argument (the calling Series or DataFrame) and that returns valid output for indexing (one of the above).

# In the given example: data.loc[0]
# It accesses the row with the index label 0.  If the index of the DataFrame 'data' is the default integer index (0, 1, 2, ...), then this will retrieve the first row.  However, if the index has been modified, then it will retrieve the row whose index label is equal to 0.
```

```python
#iloc
# .iloc is primarily integer position based (from 0 to length-1 of the axis), but may also be used with a boolean array.
# .iloc will raise IndexError if a requested indexer is out-of-bounds, except slice indexers which allow out-of-bounds indexing.

# Allowed inputs are:
# An integer, e.g. 5.
# A list or array of integers, e.g. [4, 3, 0].
# A slice object with ints, e.g. 1:7.
# A boolean array.
# A callable function with one argument (the calling Series or DataFrame) and that returns valid output for indexing (one of the above).

# In the given example: data.iloc[1]
# It accesses the second row of the DataFrame 'data' (index position 1).  Unlike .loc which is label-based, .iloc is purely position based. It always refers to the position in the DataFrame regardless of the index labels. So, data.iloc[1] will always return the second row, no matter what the index labels are.
```

```python
import pandas as pd
import numpy as np

# Sample DataFrame (replace with your actual data loading)
data = pd.DataFrame({'A': [1, 2, 3, 4, 5], 'B': [6, 7, 8, 9, 10]}, index=['a', 'b', 'c', 'd', 'e'])

# .loc
print("Using .loc:")
print(data.loc['a'])  # Access row with label 'a'
print(data.loc[['a', 'c']])  # Access rows with labels 'a' and 'c'
print(data.loc['a':'c']) # Access rows with labels from 'a' to 'c' (inclusive)

# .iloc
print("\nUsing .iloc:")
print(data.iloc[0])  # Access the first row (index position 0)
print(data.iloc[[0, 2]])  # Access the first and third rows
print(data.iloc[0:2])  # Access the first and second rows

# Demonstrating the difference with a non-default index:

# Change the index to something non-default
data.index = [10, 20, 30, 40, 50]

print("\nDataFrame with modified index:")
print(data)

print("\nUsing .loc (label-based) with modified index:")
print(data.loc[10]) # Access row with label 10

print("\nUsing .iloc (position-based) with modified index:")
print(data.iloc[0])  # Still access the first row, irrespective of the index label


#Illustrating key differences using boolean indexing
print("\nBoolean indexing with .loc:")
print(data.loc[data['A'] > 2]) # Selects rows where column 'A' is greater than 2

print("\nBoolean indexing with .iloc:")
#Direct boolean indexing with .iloc isn't possible. You'd need to find the index positions first
print(data.iloc[np.where(data['A'] > 2)[0]])
```

```python
# prompt:  graph for difference between  loc and iloc

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Sample DataFrame
data = pd.DataFrame({'A': [1, 2, 3, 4, 5], 'B': [6, 7, 8, 9, 10]}, index=['a', 'b', 'c', 'd', 'e'])

# Create a figure and axes
fig, ax = plt.subplots(figsize=(10, 6))

# Set the title and labels
ax.set_title("loc vs. iloc")
ax.set_xlabel("Indexing Method")
ax.set_ylabel("Row Access")


# Create the bar graph for loc
loc_heights = [1, 1,1]
loc_labels = ['Label-based', 'Boolean Indexing', 'Slicing with labels']
loc_x_pos = np.arange(len(loc_labels))
loc_bars = ax.bar(loc_x_pos - 0.2, loc_heights, width=0.4, label="loc", color='skyblue')

# Add value labels for loc
for bar, label in zip(loc_bars, loc_heights):
    height = bar.get_height()
    ax.text(bar.get_x() + bar.get_width() / 2, height, str(label), ha='center', va='bottom')


# Create the bar graph for iloc
iloc_heights = [1, 0, 1] #boolean indexing not directly supported
iloc_labels = ['Integer-based', 'Boolean Indexing', 'Slicing with integers']
iloc_x_pos = np.arange(len(iloc_labels))
iloc_bars = ax.bar(iloc_x_pos + 0.2, iloc_heights, width=0.4, label="iloc", color='lightcoral')


# Add value labels for iloc
for bar, label in zip(iloc_bars, iloc_heights):
    height = bar.get_height()
    ax.text(bar.get_x() + bar.get_width() / 2, height, str(label), ha='center', va='bottom')

# Set the x-axis ticks and labels
ax.set_xticks(iloc_x_pos)
ax.set_xticklabels(iloc_labels, rotation=45, ha='right') #Rotate labels for better readability

# Add legend and grid
ax.legend()
ax.grid(axis='y', linestyle='--', alpha=0.7)

plt.tight_layout()  # Adjust layout to prevent labels from overlapping
plt.show()
```

```python
# Difference between loc and iloc

#The code demonstrates the key differences between `.loc` and `.iloc` for accessing data in Pandas DataFrames.  Let's break down each part:

#**1. `.loc` (label-based indexing):**

#* **Primary Use:** Accessing data using row and column *labels*.  Think of it as looking up data using names or identifiers.
#* **Behavior:**
 #   * Uses labels in the index and column names.
 #   * *Inclusive* when slicing with labels (e.g., `'a':'c'` includes both rows 'a' and 'c').
  #  * Raises a `KeyError` if a label is not found.
#* **Examples:**
 ##  * `data.loc[['a', 'c']]`: Gets rows with index labels 'a' and 'c'.
   # * `data.loc['a':'c']`: Gets rows from index label 'a' to 'c' (inclusive).
    #* Boolean indexing: `data.loc[data['A'] > 2]`: Selects rows where the value in column 'A' is greater than 2.  This is a very common and powerful way to filter data.

#**2. `.iloc` (integer-position based indexing):**

#* **Primary Use:** Accessing data using integer *positions*.  Think of it as accessing data by its numerical location in the DataFrame.
#* **Behavior:**
#    * Uses integer positions (0-based indexing).
 #   * *Exclusive* of the upper bound when slicing (e.g., `0:2` gets rows 0 and 1, not 2).
#    * Raises an `IndexError` if the position is out of bounds (except for slices, which are more forgiving).
#* **Examples:**
#    * `data.iloc[0]`: Gets the first row (position 0).
 #   * `data.iloc[[0, 2]]`: Gets the first and third rows (positions 0 and 2).
 #   * `data.iloc[0:2]`: Gets the first and second rows (positions 0 and 1).  Note the exclusion of position 2.
 #   * Boolean indexing with .iloc isn't straightforward.  You need to first use `np.where` to get the index positions.

#**3.  Index Modification:**

#The code demonstrates that `.loc` and `.iloc` behave differently when the DataFrame's index is modified from the default integer index. `.iloc` still refers to the row positions, while `.loc` uses the new labels.

#**4. Boolean Indexing:**

#Boolean indexing with `.loc` is directly applicable using conditions, like `data['A'] > 2`.  With `.iloc`, you need to get the actual integer indices first using `np.where()` before applying it.

#**In summary:**

#Use `.loc` when you know the row and/or column *labels*.  Use `.iloc` when you know the integer *position* of the rows and/or columns.  The key difference is how they handle the DataFrame indexâ€”`.loc` uses labels while `.iloc` uses positions.  Choosing the right method ensures clarity and prevents unexpected results, especially when dealing with DataFrames that have non-default indices.
```

```python
#AXIS[0]
# axis=0 means the operation is applied row-wise (along the rows).
# For example, when dropping rows, axis=0 refers to removing rows from the DataFrame.
# If you're performing operations like summing or aggregating, it will compute these operations for each column across rows (i.e., it aggregates the data down the rows).
```

```python
import pandas as pd

# Sample DataFrame (replace with your actual data)
data = {'A': [1, 2, 3], 'B': [4, 5, 6], 'C': [7, 8, 9]}
df = pd.DataFrame(data)

# 1. Dropping a column by name using 'drop':
df_dropped = df.drop('B', axis=1)  # Drop column 'B'
print("Dropping column 'B':\n", df_dropped)

# 2. Dropping multiple columns by name:
df_dropped_multiple = df.drop(['A', 'C'], axis=1)  # Drop columns 'A' and 'C'
print("\nDropping columns 'A' and 'C':\n", df_dropped_multiple)

# 3. Dropping a column by index position using 'iloc' and 'drop':
df_dropped_by_index = df.drop(df.columns[1], axis=1)  # Drop the second column (index 1)
print("\nDropping the second column by index:\n", df_dropped_by_index)

# 4. Dropping a column in-place (modifying the original DataFrame):
df.drop('B', axis=1, inplace=True)  # Drop column 'B' directly from 'df'
print("\nOriginal DataFrame after dropping 'B' in-place:\n", df)
```

```python
#AXIS[1]
# axis=1 refers to operations along columns.
# When you specify axis=1, you're telling pandas to perform the operation on the columns (instead of the rows).
# For example, when dropping columns, axis=1 indicates that columns should be removed.
# When performing aggregation functions like sum or mean, axis=1 calculates these operations across rows for each column.
```

```python
import pandas as pd

data = {'A': [1, 2, 3], 'B': [4, 5, 6]}
df = pd.DataFrame(data)

# Dropping a row (axis=0)
df_dropped_row = df.drop(0, axis=0)  # Remove row with index 0
print("DataFrame after dropping row:\n", df_dropped_row)

# Dropping a column (axis=1)
df_dropped_col = df.drop('B', axis=1)  # Remove column named 'B'
print("\nDataFrame after dropping column:\n", df_dropped_col)
```

```python
# prompt: graph for  difference between axis[0] and axis[1] in dropping

import pandas as pd
import matplotlib.pyplot as plt

# Sample DataFrame (replace with your actual data)
data = {'A': [1, 2, 3], 'B': [4, 5, 6]}
df = pd.DataFrame(data)

# Dropping a row (axis=0)
df_dropped_row = df.drop(0, axis=0)  # Remove row with index 0
# Dropping a column (axis=1)
df_dropped_col = df.drop('B', axis=1)  # Remove column named 'B'

# Data for the bar plot
categories = ['Dropping Row', 'Dropping Column']
axis_0_values = [df.shape[0] - df_dropped_row.shape[0], 0] # axis 0 is row
axis_1_values = [0, df.shape[1] - df_dropped_col.shape[1]] # axis 1 is column


# Create the bar plot
plt.figure(figsize=(8, 6))
bar_width = 0.35

# Plotting the bars for axis=0 and axis=1
plt.bar(categories, axis_0_values, bar_width, label='axis=0 (rows)', color='skyblue')
plt.bar([i for i in categories], axis_1_values, bar_width, label='axis=1 (columns)', bottom=axis_0_values, color='lightcoral')


# Adding labels and title
plt.xlabel("Operation")
plt.ylabel("Difference in shape")
plt.title("Difference between axis=0 and axis=1 when dropping")

# Adding value labels on top of the bars
for i, v in enumerate(axis_0_values):
    plt.text(i, v / 2, str(v), ha='center', va='center', color='black')

for i, v in enumerate(axis_1_values):
    plt.text(i, v / 2 + axis_0_values[i], str(v), ha='center', va='center', color='black')


# Adding legend
plt.legend()

# Show the plot
plt.show()
```
